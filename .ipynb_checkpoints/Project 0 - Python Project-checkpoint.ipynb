### المهمة الثالثة

قم بكتابة دالة تقوم بمقارنة الرقم المدخل (مثل ١٩) بالرقم ١٠ و تخبرك ما إذا كان أكبر من الرقم ١٠ أم لا

```python
def bigger_than_10(x):
    if(x>10):
        print("الرقم الذي أدخلته أكبر من 10")
    else:
        print("  الرقم الذي أدخلته ليس أكبر  من 10 ")
bigger_than_10(19)
```

### المهمة الخامسة

لنفترض أن لدينا المعجم التالي، نريد إضافة المفتاح  
black  
بالقيمة  
000000

```python
colors = {
  "blue": "0000FF",
  "green": "00FF00",
  "yellow": "FFFF00",
  "red": "FF0000",
  "white": "unknown"
}
```

```python
colors["black"]="000000"
```

هل يمكننا تغيير قيمة المفتاح  
white  
بالقيمة  
FFFFFF ?

```python
colors["white"]="FFFFFF"
```

نريد إنشاء دالة تقوم بأخذ قائمة و استبدال عناصرها بالقيم من المعجم بالاستعانة بالمفاتيح
مثال: نستبدل  
blue  
  بالقيمة  
  0000FF

```python
def exchange_values(lista):
    for count,ele in enumerate(lista):
            lista[count]=colors[ele]
    return lista
    
```

```python
lista = ['blue', 'white', 'black', 'yellow', 'green', 'red']
exchange_values(lista)
```
